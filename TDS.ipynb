### KMeans Clustering

```python
import numpy as np
import pandas as pd

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('/content/stock.csv')
```

```python
df.head()
```

```python
df1 = df.copy(deep=True)
```

```python
df1.drop('record number', inplace=True, axis=1)
```

```python
df1.head()
```

```python
np.random.seed(1729)
kmeans =  KMeans(n_clusters=7, random_state=1729)
clus = kmeans.fit_predict(df1)
```

```python
df['cluster'] = clus
```

```python
df.head()
```

```python
df[df['record number'] == 0].iloc[0][-1]
```

```python
df['cluster'].value_counts()
```

```python
df[df['record number'] == 1].iloc[0][-1]
```

```python
records = [(651,752),(593,181),(510,213),(115,819)]

for rec in records:
    r1,r2 = rec
    clus1 = df[df['record number'] == r1].iloc[0][-1]
    clus2 = df[df['record number'] == r2].iloc[0][-1]
    print(f"Record ({r1},{r2}) has center ({clus1},{clus2})")
    if clus1 == clus2:
        print("------"*5)
        print("The same cluster ceter is",clus1)
        print("------"*5)
```

### Live Coding

```python
#common imports
import numpy as np
import pandas as pd

#import for geospatial analysis
import geopy.distance

#import for visualizing maps
import folium
```

```python
#@title Live Coding

url = 'https://drive.google.com/uc?id=19JERBe0ndadkMPKDFZ0rTEZqdI50zDfk'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
#Let us use folium to visualize the data on google maps

empb = [40.748, -73.98572]
folium.Map(location=empb, zoom_start=10)
```

```python
df['coordinates'] = "(" + df['lat'].astype(str) + "," + df['lng'].astype(str) + ")"
```

```python
df.head()
```

```python
empb = [40.748, -73.98572]
map = folium.Map(location=empb, zoom_start=10)

for i,row in df.iterrows():
    # if i==10:
    #     break
    # print(row[0])
    # print(i, "---------------")
    lat = df.loc[i,'lat']
    lng = df.loc[i,'lng']
    store = df.loc[i,'store']

    if store == "Starbucks":
        color = "green"
    else:
        color = 'red'

    folium.Marker(location=[lat,lng], popup=store, icon=folium.Icon(color)).add_to(map) #takes color to denote which store it is

map
```

```python
distance = []

for row in df.itertuples(): #if you have dilemma then use itertuples
    dist = geopy.distance.geodesic(empb, row.coordinates).km
    distance.append(dist)
```

```python
distance[:20]
```

```python
df['distance'] = distance
```

```python
df.head()
```

```python
df[df['distance'] > 10]
```

```python
df.groupby('store').agg({'distance':['mean', 'max', 'min']})
```
